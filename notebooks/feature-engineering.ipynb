```python
# %% [markdown]
# # Feature Engineering for Fraud Detection
# 
# This notebook focuses on creating and evaluating features for both e-commerce and credit card fraud detection datasets.
# 
# ## Objectives:
# 1. Create temporal features from timestamps
# 2. Engineer behavioral pattern features
# 3. Create risk indicators and aggregated features
# 4. Transform and encode features
# 5. Evaluate feature importance and correlation

# %% [markdown]
# ## 1. Setup and Imports

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Visualization settings
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12

# Import custom modules
import sys
sys.path.append('../src')

try:
    from feature_engineering import FeatureEngineer
    from data_preprocessing import DataPreprocessor
    from utils import setup_logging
except ImportError:
    print("Custom modules not found. Creating simplified versions...")
    
    # Create simplified versions if modules don't exist
    class FeatureEngineer:
        def create_time_features(self, df):
            """Create temporal features from timestamp columns."""
            df = df.copy()
            
            # Convert to datetime if not already
            if 'purchase_time' in df.columns:
                df['purchase_time'] = pd.to_datetime(df['purchase_time'])
            if 'signup_time' in df.columns:
                df['signup_time'] = pd.to_datetime(df['signup_time'])
            
            # Extract time components
            if 'purchase_time' in df.columns:
                df['purchase_hour'] = df['purchase_time'].dt.hour
                df['purchase_dayofweek'] = df['purchase_time'].dt.dayofweek
                df['purchase_day'] = df['purchase_time'].dt.day
                df['purchase_month'] = df['purchase_time'].dt.month
                
                # Business hours (9 AM to 5 PM)
                df['is_business_hours'] = df['purchase_hour'].between(9, 17).astype(int)
                # Weekend flag
                df['is_weekend'] = df['purchase_dayofweek'].isin([5, 6]).astype(int)
            
            if 'signup_time' in df.columns:
                df['signup_hour'] = df['signup_time'].dt.hour
                df['signup_dayofweek'] = df['signup_time'].dt.dayofweek
            
            # Time since signup
            if all(col in df.columns for col in ['purchase_time', 'signup_time']):
                df['time_since_signup'] = (df['purchase_time'] - df['signup_time']).dt.total_seconds() / 3600
            
            return df
        
        def create_transaction_features(self, df):
            """Create transaction pattern features."""
            df = df.copy()
            
            # Transaction velocity (transactions per hour per user)
            if 'user_id' in df.columns and 'purchase_time' in df.columns:
                df_sorted = df.sort_values(['user_id', 'purchase_time'])
                df_sorted['time_diff'] = df_sorted.groupby('user_id')['purchase_time'].diff().dt.total_seconds() / 3600
                df_sorted['time_diff'].fillna(1, inplace=True)  # First transaction
                df_sorted['transaction_velocity'] = 1 / df_sorted['time_diff']
                df = df_sorted.drop('time_diff', axis=1)
            
            # User transaction statistics
            if 'user_id' in df.columns:
                # Total transactions per user
                df['user_total_transactions'] = df.groupby('user_id')['user_id'].transform('count')
                
                # Purchase value statistics
                if 'purchase_value' in df.columns:
                    df['user_purchase_mean'] = df.groupby('user_id')['purchase_value'].transform('mean')
                    df['user_purchase_std'] = df.groupby('user_id')['purchase_value'].transform('std')
                    df['user_purchase_max'] = df.groupby('user_id')['purchase_value'].transform('max')
                    df['user_purchase_min'] = df.groupby('user_id')['purchase_value'].transform('min')
                    
                    # Purchase value deviation (handle NaN from std)
                    df['purchase_value_deviation'] = (df['purchase_value'] - df['user_purchase_mean'])
                    # Only divide by std if it's not 0 or NaN
                    mask = df['user_purchase_std'].notna() & (df['user_purchase_std'] != 0)
                    df.loc[mask, 'purchase_value_deviation'] = df.loc[mask, 'purchase_value_deviation'] / df.loc[mask, 'user_purchase_std']
                    df.loc[~mask, 'purchase_value_deviation'] = 0
            
            # Device and IP sharing
            if 'device_id' in df.columns:
                df['unique_users_per_device'] = df.groupby('device_id')['user_id'].transform('nunique')
            
            if 'ip_address' in df.columns:
                df['unique_users_per_ip'] = df.groupby('ip_address')['user_id'].transform('nunique')
            
            return df
        
        def create_browser_source_features(self, df):
            """Create browser and source risk features."""
            df = df.copy()
            
            # Browser fraud rate
            if 'browser' in df.columns and 'class' in df.columns:
                browser_stats = df.groupby('browser')['class'].agg(['mean', 'count']).reset_index()
                browser_stats.rename(columns={'mean': 'browser_fraud_rate', 'count': 'browser_count'}, inplace=True)
                df = pd.merge(df, browser_stats[['browser', 'browser_fraud_rate']], on='browser', how='left')
            
            # Source fraud rate
            if 'source' in df.columns and 'class' in df.columns:
                source_stats = df.groupby('source')['class'].agg(['mean', 'count']).reset_index()
                source_stats.rename(columns={'mean': 'source_fraud_rate', 'count': 'source_count'}, inplace=True)
                df = pd.merge(df, source_stats[['source', 'source_fraud_rate']], on='source', how='left')
            
            # Country fraud rate
            if 'country' in df.columns and 'class' in df.columns:
                country_stats = df.groupby('country')['class'].agg(['mean', 'count']).reset_index()
                country_stats.rename(columns={'mean': 'country_fraud_rate', 'count': 'country_count'}, inplace=True)
                df = pd.merge(df, country_stats[['country', 'country_fraud_rate']], on='country', how='left')
            
            # Browser-source combination fraud rate
            if all(col in df.columns for col in ['browser', 'source', 'class']):
                df['browser_source'] = df['browser'] + '_' + df['source']
                combo_stats = df.groupby('browser_source')['class'].agg(['mean', 'count']).reset_index()
                combo_stats.rename(columns={'mean': 'browser_source_fraud_rate', 'count': 'browser_source_count'}, inplace=True)
                df = pd.merge(df, combo_stats[['browser_source', 'browser_source_fraud_rate']], on='browser_source', how='left')
                df.drop('browser_source', axis=1, inplace=True)
            
            return df
        
        def encode_categorical_features(self, df, categorical_cols):
            """Encode categorical features."""
            df = df.copy()
            
            for col in categorical_cols:
                if col in df.columns:
                    # Frequency encoding for high cardinality features
                    if df[col].nunique() > 20:
                        freq_map = df[col].value_counts(normalize=True).to_dict()
                        df[f'{col}_freq'] = df[col].map(freq_map)
                        df.drop(col, axis=1, inplace=True)
                    else:
                        # One-hot encoding for low cardinality features
                        dummies = pd.get_dummies(df[col], prefix=col, drop_first=True)
                        df = pd.concat([df.drop(col, axis=1), dummies], axis=1)
            
            return df
        
        def scale_numerical_features(self, df, numerical_cols):
            """Scale numerical features using robust scaling."""
            df = df.copy()
            
            from sklearn.preprocessing import RobustScaler
            
            scaler = RobustScaler()
            
            for col in numerical_cols:
                if col in df.columns:
                    # Handle NaN values before scaling
                    if df[col].isna().any():
                        df[col] = df[col].fillna(df[col].median())
                    
                    # Scale the column
                    df[col] = scaler.fit_transform(df[[col]])
            
            return df
        
        def handle_missing_values(self, df):
            """Handle missing values in the dataset."""
            df = df.copy()
            
            # For numerical columns, fill with median
            num_cols = df.select_dtypes(include=[np.number]).columns
            for col in num_cols:
                if df[col].isna().any():
                    df[col] = df[col].fillna(df[col].median())
            
            # For categorical columns, fill with mode
            cat_cols = df.select_dtypes(include=['object']).columns
            for col in cat_cols:
                if df[col].isna().any():
                    df[col] = df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else 'Unknown')
            
            return df
        
        def prepare_creditcard_features(self, df):
            """Prepare features for credit card dataset."""
            df = df.copy()
            
            # Time-based features
            if 'Time' in df.columns:
                # Convert seconds to hours and days
                df['hour_of_day'] = (df['Time'] // 3600) % 24
                df['day_of_week'] = (df['Time'] // (3600 * 24)) % 7
            
            # Amount transformation
            if 'Amount' in df.columns:
                df['Amount_log'] = np.log1p(df['Amount'])
            
            # V features statistics
            v_columns = [col for col in df.columns if col.startswith('V')]
            if v_columns:
                df['V_mean'] = df[v_columns].mean(axis=1)
                df['V_std'] = df[v_columns].std(axis=1)
                df['V_max'] = df[v_columns].max(axis=1)
                df['V_min'] = df[v_columns].min(axis=1)
                
                # Interaction terms
                if 'V1' in df.columns and 'Amount_log' in df.columns:
                    df['V1_Amount_interaction'] = df['V1'] * df['Amount_log']
            
            return df

    class DataPreprocessor:
        def load_data(self, fraud_path, ip_path, creditcard_path):
            """Load all datasets."""
            fraud_data = pd.read_csv(fraud_path)
            ip_country = pd.read_csv(ip_path)
            creditcard = pd.read_csv(creditcard_path)
            return fraud_data, ip_country, creditcard
        
        def clean_fraud_data(self, df):
            """Clean e-commerce fraud data."""
            df = df.copy()
            
            # Convert timestamps
            if 'purchase_time' in df.columns:
                df['purchase_time'] = pd.to_datetime(df['purchase_time'])
            if 'signup_time' in df.columns:
                df['signup_time'] = pd.to_datetime(df['signup_time'])
            
            # Handle missing values
            df = df.dropna(subset=['user_id', 'purchase_value'])
            
            return df
        
        def map_ip_to_country(self, df, ip_country):
            """Map IP addresses to countries."""
            return df  # Simplified version
        
        def clean_creditcard_data(self, df):
            """Clean credit card data."""
            df = df.copy()
            
            # Remove duplicates
            df = df.drop_duplicates()
            
            # Handle any potential NaN values
            df = df.fillna(0)
            
            return df
    
    def setup_logging():
        """Setup logging configuration."""
        import logging
        logging.basicConfig(level=logging.INFO)
        
    print("Using simplified custom modules.")

print("Libraries imported successfully!")

# %% [markdown]
# ## 2. Load and Prepare Data

# %%
# Initialize preprocessor and load data
preprocessor = DataPreprocessor()

# Load all datasets (using sample data if files don't exist)
print("Loading datasets...")
try:
    fraud_data, ip_country, creditcard = preprocessor.load_data(
        '../data/raw/Fraud_Data.csv',
        '../data/raw/IpAddress_to_Country.csv',
        '../data/raw/creditcard.csv'
    )
except FileNotFoundError:
    print("Data files not found. Creating sample data for demonstration...")
    
    # Create sample e-commerce data
    n_samples = 10000
    fraud_data = pd.DataFrame({
        'user_id': range(1, n_samples + 1),
        'signup_time': pd.date_range('2023-01-01', periods=n_samples, freq='H'),
        'purchase_time': pd.date_range('2023-01-01', periods=n_samples, freq='H') + pd.Timedelta(hours=np.random.randint(1, 168, n_samples)),
        'purchase_value': np.random.exponential(100, n_samples),
        'device_id': np.random.choice(['device_' + str(i) for i in range(1, 101)], n_samples),
        'source': np.random.choice(['Direct', 'SEO', 'Ads'], n_samples),
        'browser': np.random.choice(['Chrome', 'Firefox', 'Safari', 'Edge'], n_samples),
        'sex': np.random.choice(['M', 'F'], n_samples),
        'age': np.random.randint(18, 70, n_samples),
        'ip_address': ['192.168.' + str(np.random.randint(1, 255)) + '.' + str(np.random.randint(1, 255)) for _ in range(n_samples)],
        'class': np.random.binomial(1, 0.01, n_samples)  # 1% fraud rate
    })
    
    # Create sample credit card data
    n_cc_samples = 10000
    creditcard = pd.DataFrame({
        'Time': np.random.randint(0, 172800, n_cc_samples),  # 48 hours in seconds
        'Amount': np.random.exponential(100, n_cc_samples),
        'Class': np.random.binomial(1, 0.0017, n_cc_samples)  # 0.17% fraud rate (typical)
    })
    
    # Add PCA-like features
    for i in range(1, 29):
        creditcard[f'V{i}'] = np.random.normal(0, 1, n_cc_samples)
    
    ip_country = pd.DataFrame()  # Empty for sample data

print(f"E-commerce data shape: {fraud_data.shape}")
print(f"Credit card data shape: {creditcard.shape}")

# Clean data
fraud_clean = preprocessor.clean_fraud_data(fraud_data)
fraud_with_country = preprocessor.map_ip_to_country(fraud_clean, ip_country)
cc_clean = preprocessor.clean_creditcard_data(creditcard)

print("\nData cleaning completed!")
print(f"E-commerce cleaned shape: {fraud_with_country.shape}")
print(f"Credit card cleaned shape: {cc_clean.shape}")

# %% [markdown]
# ## 3. E-commerce Data Feature Engineering

# %%
# Initialize feature engineer
feature_engineer = FeatureEngineer()

print("="*70)
print("E-COMMERCE DATA FEATURE ENGINEERING")
print("="*70)

# %% [markdown]
# ### 3.1 Original Data Overview

# %%
print("Original Features in E-commerce Data:")
print("-" * 50)
for i, col in enumerate(fraud_with_country.columns, 1):
    print(f"{i:2}. {col:20} | Type: {fraud_with_country[col].dtype} | Unique: {fraud_with_country[col].nunique():,}")

# Display sample of original data
print("\nSample of original data (first 3 rows):")
display(fraud_with_country.head(3))

# %% [markdown]
# ### 3.2 Create Temporal Features

# %%
print("\nCreating temporal features...")

# Create time-based features
fraud_temporal = feature_engineer.create_time_features(fraud_with_country)

# Display new temporal features
temporal_features = [col for col in fraud_temporal.columns if col not in fraud_with_country.columns]
print(f"\nCreated {len(temporal_features)} temporal features:")
for i, feat in enumerate(temporal_features, 1):
    print(f"{i:2}. {feat}")

# Visualize temporal features
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

# Plot purchase hour distribution
axes[0].hist(fraud_temporal['purchase_hour'].dropna(), bins=24, alpha=0.7, color='#2E75B6', edgecolor='black')
axes[0].set_title('Purchase Hour Distribution', fontweight='bold')
axes[0].set_xlabel('Hour of Day (0-23)')
axes[0].set_ylabel('Frequency')
axes[0].grid(True, alpha=0.3)

# Plot day of week distribution
day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
day_counts = fraud_temporal['purchase_dayofweek'].value_counts().sort_index()
axes[1].bar(day_names, day_counts.values, color='#2E75B6', alpha=0.7)
axes[1].set_title('Purchase Day of Week Distribution', fontweight='bold')
axes[1].set_xlabel('Day of Week')
axes[1].set_ylabel('Count')
axes[1].grid(True, alpha=0.3, axis='y')

# Plot time since signup distribution (log scale)
time_since_signup_hours = fraud_temporal['time_since_signup'].dropna()
axes[2].hist(time_since_signup_hours[time_since_signup_hours < 168], bins=50, 
             alpha=0.7, color='#2E75B6', edgecolor='black')  # First 7 days
axes[2].set_title('Time Since Signup (First 7 Days)', fontweight='bold')
axes[2].set_xlabel('Hours Since Signup')
axes[2].set_ylabel('Frequency')
axes[2].grid(True, alpha=0.3)

# Plot business hours vs non-business hours
business_hours_counts = fraud_temporal['is_business_hours'].value_counts()
axes[3].pie(business_hours_counts.values, labels=['Non-Business', 'Business Hours'],
           colors=['#FF6B6B', '#4ECDC4'], autopct='%1.1f%%', startangle=90)
axes[3].set_title('Business Hours vs Non-Business Hours', fontweight='bold')

# Plot weekend vs weekday
weekend_counts = fraud_temporal['is_weekend'].value_counts()
axes[4].pie(weekend_counts.values, labels=['Weekday', 'Weekend'],
           colors=['#45B7D1', '#96CEB4'], autopct='%1.1f%%', startangle=90)
axes[4].set_title('Weekday vs Weekend Transactions', fontweight='bold')

# Plot signup hour distribution
axes[5].hist(fraud_temporal['signup_hour'].dropna(), bins=24, alpha=0.7, color='#2E75B6', edgecolor='black')
axes[5].set_title('Signup Hour Distribution', fontweight='bold')
axes[5].set_xlabel('Hour of Day (0-23)')
axes[5].set_ylabel('Frequency')
axes[5].grid(True, alpha=0.3)

plt.tight_layout()
plt.suptitle('Temporal Feature Distributions - E-commerce Data', fontsize=14, fontweight='bold', y=1.02)
plt.show()

# %% [markdown]
# ### 3.3 Create Transaction Pattern Features

# %%
print("\nCreating transaction pattern features...")

# Create transaction features
fraud_transaction = feature_engineer.create_transaction_features(fraud_temporal)

# Display new transaction features
transaction_features = [col for col in fraud_transaction.columns if col not in fraud_temporal.columns]
print(f"\nCreated {len(transaction_features)} transaction pattern features:")
for i, feat in enumerate(transaction_features, 1):
    print(f"{i:2}. {feat}")

# Visualize key transaction features
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# 1. Transaction velocity distribution
axes[0, 0].hist(fraud_transaction['transaction_velocity'].clip(upper=10), bins=50, 
                alpha=0.7, color='#2E75B6', edgecolor='black')
axes[0, 0].set_title('Transaction Velocity (Clipped at 10)', fontweight='bold')
axes[0, 0].set_xlabel('Transactions per Hour')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(True, alpha=0.3)

# 2. User total transactions
axes[0, 1].hist(fraud_transaction['user_total_transactions'].clip(upper=20), bins=20, 
                alpha=0.7, color='#2E75B6', edgecolor='black')
axes[0, 1].set_title('User Total Transactions (Clipped at 20)', fontweight='bold')
axes[0, 1].set_xlabel('Number of Transactions')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].grid(True, alpha=0.3)

# 3. Unique users per device (if column exists)
if 'unique_users_per_device' in fraud_transaction.columns:
    device_counts = fraud_transaction['unique_users_per_device'].value_counts().head(10)
    axes[0, 2].bar(range(len(device_counts)), device_counts.values, 
                   color=['#FF6B6B' if x > 1 else '#4ECDC4' for x in device_counts.index])
    axes[0, 2].set_title('Top 10: Unique Users per Device', fontweight='bold')
    axes[0, 2].set_xlabel('Number of Unique Users')
    axes[0, 2].set_ylabel('Device Count')
    axes[0, 2].set_xticks(range(len(device_counts)))
    axes[0, 2].set_xticklabels(device_counts.index)
    axes[0, 2].grid(True, alpha=0.3, axis='y')
else:
    axes[0, 2].text(0.5, 0.5, 'Device ID not available', 
                   ha='center', va='center', fontsize=12)
    axes[0, 2].set_title('Unique Users per Device', fontweight='bold')
    axes[0, 2].axis('off')

# 4. Unique users per IP (if column exists)
if 'unique_users_per_ip' in fraud_transaction.columns:
    ip_counts = fraud_transaction['unique_users_per_ip'].value_counts().head(10)
    axes[1, 0].bar(range(len(ip_counts)), ip_counts.values, 
                   color=['#FF6B6B' if x > 1 else '#4ECDC4' for x in ip_counts.index])
    axes[1, 0].set_title('Top 10: Unique Users per IP', fontweight='bold')
    axes[1, 0].set_xlabel('Number of Unique Users')
    axes[1, 0].set_ylabel('IP Count')
    axes[1, 0].set_xticks(range(len(ip_counts)))
    axes[1, 0].set_xticklabels(ip_counts.index)
    axes[1, 0].grid(True, alpha=0.3, axis='y')
else:
    axes[1, 0].text(0.5, 0.5, 'IP Address not available', 
                   ha='center', va='center', fontsize=12)
    axes[1, 0].set_title('Unique Users per IP', fontweight='bold')
    axes[1, 0].axis('off')

# 5. Purchase value deviation (check for NaN values)
if 'purchase_value_deviation' in fraud_transaction.columns:
    # Handle NaN values before plotting
    deviation_data = fraud_transaction['purchase_value_deviation'].dropna()
    if len(deviation_data) > 0:
        axes[1, 1].hist(deviation_data.clip(lower=-5, upper=5), bins=50, 
                        alpha=0.7, color='#2E75B6', edgecolor='black')
        axes[1, 1].set_title('Purchase Value Deviation from User Mean', fontweight='bold')
        axes[1, 1].set_xlabel('Standard Deviations from Mean')
        axes[1, 1].set_ylabel('Frequency')
        axes[1, 1].grid(True, alpha=0.3)
    else:
        axes[1, 1].text(0.5, 0.5, 'No valid deviation data', 
                       ha='center', va='center', fontsize=12)
        axes[1, 1].set_title('Purchase Value Deviation', fontweight='bold')
        axes[1, 1].axis('off')
else:
    axes[1, 1].text(0.5, 0.5, 'Purchase value deviation\nnot calculated', 
                   ha='center', va='center', fontsize=12)
    axes[1, 1].set_title('Purchase Value Deviation', fontweight='bold')
    axes[1, 1].axis('off')

# 6. User purchase statistics correlation
stats_cols = ['user_purchase_mean', 'user_purchase_std', 'user_purchase_max', 'user_purchase_min']
available_stats = [col for col in stats_cols if col in fraud_transaction.columns]

if len(available_stats) >= 2:
    user_stats = fraud_transaction[available_stats].corr()
    im = axes[1, 2].imshow(user_stats, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
    axes[1, 2].set_title('User Purchase Statistics Correlation', fontweight='bold')
    axes[1, 2].set_xticks(range(len(available_stats)))
    axes[1, 2].set_yticks(range(len(available_stats)))
    axes[1, 2].set_xticklabels([col.replace('user_purchase_', '') for col in available_stats], rotation=45)
    axes[1, 2].set_yticklabels([col.replace('user_purchase_', '') for col in available_stats])
    plt.colorbar(im, ax=axes[1, 2])
else:
    axes[1, 2].text(0.5, 0.5, 'Insufficient purchase statistics\nfor correlation matrix', 
                   ha='center', va='center', fontsize=12)
    axes[1, 2].set_title('User Purchase Statistics', fontweight='bold')
    axes[1, 2].axis('off')

plt.tight_layout()
plt.suptitle('Transaction Pattern Features - E-commerce Data', fontsize=14, fontweight='bold', y=1.02)
plt.show()

# %% [markdown]
# ### 3.4 Create Browser and Source Risk Features

# %%
print("\nCreating browser and source risk features...")

# Create browser/source features
fraud_browser_source = feature_engineer.create_browser_source_features(fraud_transaction)

# Display new browser/source features
browser_source_features = [col for col in fraud_browser_source.columns if col not in fraud_transaction.columns]
print(f"\nCreated {len(browser_source_features)} browser/source risk features:")
for i, feat in enumerate(browser_source_features, 1):
    print(f"{i:2}. {feat}")

# Visualize browser and source risk features
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Browser fraud rates
if 'browser_fraud_rate' in fraud_browser_source.columns:
    browser_fraud_rates = fraud_browser_source.groupby('browser')['browser_fraud_rate'].first().sort_values(ascending=False)
    colors = plt.cm.RdYlGn_r((browser_fraud_rates.values - browser_fraud_rates.values.min()) / 
                            (browser_fraud_rates.values.max() - browser_fraud_rates.values.min() + 1e-10))
    axes[0, 0].bar(range(len(browser_fraud_rates)), browser_fraud_rates.values, color=colors)
    axes[0, 0].set_title('Browser Fraud Rates', fontweight='bold', fontsize=12)
    axes[0, 0].set_xlabel('Browser')
    axes[0, 0].set_ylabel('Fraud Rate')
    axes[0, 0].set_xticks(range(len(browser_fraud_rates)))
    axes[0, 0].set_xticklabels(browser_fraud_rates.index, rotation=45, ha='right')
    axes[0, 0].grid(True, alpha=0.3, axis='y')
else:
    axes[0, 0].text(0.5, 0.5, 'Browser fraud rates\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[0, 0].set_title('Browser Fraud Rates', fontweight='bold')
    axes[0, 0].axis('off')

# 2. Source fraud rates
if 'source_fraud_rate' in fraud_browser_source.columns:
    source_fraud_rates = fraud_browser_source.groupby('source')['source_fraud_rate'].first().sort_values(ascending=False)
    colors = plt.cm.RdYlGn_r((source_fraud_rates.values - source_fraud_rates.values.min()) / 
                            (source_fraud_rates.values.max() - source_fraud_rates.values.min() + 1e-10))
    axes[0, 1].bar(range(len(source_fraud_rates)), source_fraud_rates.values, color=colors)
    axes[0, 1].set_title('Source Fraud Rates', fontweight='bold', fontsize=12)
    axes[0, 1].set_xlabel('Source')
    axes[0, 1].set_ylabel('Fraud Rate')
    axes[0, 1].set_xticks(range(len(source_fraud_rates)))
    axes[0, 1].set_xticklabels(source_fraud_rates.index, rotation=45, ha='right')
    axes[0, 1].grid(True, alpha=0.3, axis='y')
else:
    axes[0, 1].text(0.5, 0.5, 'Source fraud rates\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[0, 1].set_title('Source Fraud Rates', fontweight='bold')
    axes[0, 1].axis('off')

# 3. Country fraud rates (if available)
if 'country_fraud_rate' in fraud_browser_source.columns and 'country' in fraud_browser_source.columns:
    country_fraud_rates = fraud_browser_source.groupby('country')['country_fraud_rate'].first()
    if len(country_fraud_rates) > 0:
        top_countries = country_fraud_rates.sort_values(ascending=False).head(15)
        colors = plt.cm.RdYlGn_r((top_countries.values - top_countries.values.min()) / 
                                (top_countries.values.max() - top_countries.values.min() + 1e-10))
        axes[1, 0].bar(range(len(top_countries)), top_countries.values, color=colors)
        axes[1, 0].set_title('Top 15 Countries by Fraud Rate', fontweight='bold', fontsize=12)
        axes[1, 0].set_xlabel('Country')
        axes[1, 0].set_ylabel('Fraud Rate')
        axes[1, 0].set_xticks(range(len(top_countries)))
        axes[1, 0].set_xticklabels(top_countries.index, rotation=45, ha='right')
        axes[1, 0].grid(True, alpha=0.3, axis='y')
    else:
        axes[1, 0].text(0.5, 0.5, 'No country data available', 
                       ha='center', va='center', fontsize=12)
        axes[1, 0].set_title('Country Fraud Rates', fontweight='bold')
        axes[1, 0].axis('off')
else:
    axes[1, 0].text(0.5, 0.5, 'Country fraud rates\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[1, 0].set_title('Country Fraud Rates', fontweight='bold')
    axes[1, 0].axis('off')

# 4. Browser-source combination fraud rates
if 'browser_source_fraud_rate' in fraud_browser_source.columns and all(col in fraud_browser_source.columns for col in ['browser', 'source']):
    # Get browser and source columns
    fraud_browser_source['browser_source_combo'] = fraud_browser_source['browser'] + ' + ' + fraud_browser_source['source']
    combo_rates = fraud_browser_source.groupby('browser_source_combo')['browser_source_fraud_rate'].first()
    if len(combo_rates) > 0:
        top_combos = combo_rates.sort_values(ascending=False).head(10)
        colors = plt.cm.RdYlGn_r((top_combos.values - top_combos.values.min()) / 
                                (top_combos.values.max() - top_combos.values.min() + 1e-10))
        axes[1, 1].bar(range(len(top_combos)), top_combos.values, color=colors)
        axes[1, 1].set_title('Top 10 Browser-Source Combos by Fraud Rate', fontweight='bold', fontsize=12)
        axes[1, 1].set_xlabel('Browser + Source Combination')
        axes[1, 1].set_ylabel('Fraud Rate')
        axes[1, 1].set_xticks(range(len(top_combos)))
        axes[1, 1].set_xticklabels(top_combos.index, rotation=45, ha='right')
        axes[1, 1].grid(True, alpha=0.3, axis='y')
    else:
        axes[1, 1].text(0.5, 0.5, 'No browser-source combos\navailable', 
                       ha='center', va='center', fontsize=12)
        axes[1, 1].set_title('Browser-Source Combos', fontweight='bold')
        axes[1, 1].axis('off')
else:
    axes[1, 1].text(0.5, 0.5, 'Browser-source combo rates\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[1, 1].set_title('Browser-Source Combos', fontweight='bold')
    axes[1, 1].axis('off')

plt.tight_layout()
plt.suptitle('Risk Indicator Features - E-commerce Data', fontsize=14, fontweight='bold', y=1.02)
plt.show()

# %% [markdown]
# ### 3.5 Feature Transformation and Encoding

# %%
print("\nPerforming feature transformation and encoding...")

# Define categorical and numerical columns
categorical_cols = []
if 'source' in fraud_browser_source.columns:
    categorical_cols.append('source')
if 'browser' in fraud_browser_source.columns:
    categorical_cols.append('browser')
if 'sex' in fraud_browser_source.columns:
    categorical_cols.append('sex')
if 'country' in fraud_browser_source.columns:
    categorical_cols.append('country')

# Numerical columns
numerical_cols = []
for col in ['purchase_value', 'age', 'time_since_signup', 'transaction_velocity', 
           'user_total_transactions', 'purchase_value_deviation', 'user_purchase_mean', 
           'user_purchase_std', 'unique_users_per_device', 'unique_users_per_ip']:
    if col in fraud_browser_source.columns:
        numerical_cols.append(col)

# Add engineered numerical features
rate_features = [col for col in fraud_browser_source.columns 
                if col.endswith('_rate') and col not in numerical_cols]
numerical_cols.extend(rate_features)

print(f"\nCategorical columns to encode: {categorical_cols}")
print(f"\nNumerical columns to scale: {len(numerical_cols)} columns")

# Encode categorical features
print("\nEncoding categorical features...")
fraud_encoded = feature_engineer.encode_categorical_features(fraud_browser_source, categorical_cols)

# Scale numerical features
print("Scaling numerical features...")
fraud_scaled = feature_engineer.scale_numerical_features(fraud_encoded, numerical_cols)

# Handle missing values
print("Handling missing values...")
fraud_final = feature_engineer.handle_missing_values(fraud_scaled)

print(f"\nFinal dataset shape: {fraud_final.shape}")
print(f"Total features created: {fraud_final.shape[1] - fraud_with_country.shape[1]}")

# %% [markdown]
# ### 3.6 Feature Correlation Analysis

# %%
print("\nAnalyzing feature correlations...")

# Select numerical features for correlation analysis
numerical_features = fraud_final.select_dtypes(include=[np.number]).columns.tolist()
# Limit to top 20 features for readability
if len(numerical_features) > 20:
    numerical_features = numerical_features[:20]

# Calculate correlation matrix
corr_matrix = fraud_final[numerical_features].corr()

# Plot correlation heatmap
plt.figure(figsize=(14, 12))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .8},
            annot=True, fmt='.2f', annot_kws={'size': 8})
plt.title('Feature Correlation Heatmap - E-commerce Data (Top 20 Numerical Features)', 
          fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Find highly correlated features
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > 0.8:
            high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], 
                                   corr_matrix.iloc[i, j]))

if high_corr_pairs:
    print("\nHighly correlated feature pairs (|correlation| > 0.8):")
    for feat1, feat2, corr in high_corr_pairs:
        print(f"   {feat1}  {feat2}: {corr:.3f}")
else:
    print("\nNo highly correlated feature pairs found (|correlation| > 0.8)")

# %% [markdown]
# ### 3.7 Feature Importance Analysis (Using Correlation with Target)

# %%
print("\nAnalyzing feature importance using correlation with target...")

# Calculate correlation with target variable
if 'class' in fraud_final.columns:
    feature_importance = []
    
    for col in fraud_final.select_dtypes(include=[np.number]).columns:
        if col != 'class':
            corr = fraud_final[col].corr(fraud_final['class'])
            feature_importance.append((col, abs(corr), corr))
    
    # Sort by absolute correlation
    feature_importance.sort(key=lambda x: x[1], reverse=True)
    
    # Plot top 15 features
    top_features = feature_importance[:15]
    if top_features:  # Check if we have features
        features = [f[0] for f in top_features]
        correlations = [f[2] for f in top_features]
        abs_correlations = [f[1] for f in top_features]
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
        
        # Bar plot
        colors = ['#2E75B6' if x > 0 else '#C00000' for x in correlations]
        y_pos = np.arange(len(features))
        ax1.barh(y_pos, correlations, color=colors, alpha=0.7)
        ax1.set_yticks(y_pos)
        ax1.set_yticklabels(features)
        ax1.set_xlabel('Correlation with Fraud Target', fontsize=12)
        ax1.set_title('Top 15 Features by Correlation with Fraud', fontsize=14, fontweight='bold')
        ax1.grid(True, alpha=0.3, axis='x')
        
        # Add correlation values
        for i, (corr, feat) in enumerate(zip(correlations, features)):
            ax1.text(corr + (0.01 if corr >= 0 else -0.01), i, 
                    f'{corr:.3f}', va='center',
                    color='black', fontweight='bold')
        
        # Scatter plot of correlation vs feature
        ax2.scatter(range(len(feature_importance)), [f[1] for f in feature_importance], 
                    alpha=0.6, color='#2E75B6')
        ax2.set_xlabel('Feature Rank', fontsize=12)
        ax2.set_ylabel('Absolute Correlation', fontsize=12)
        ax2.set_title('Feature Importance Distribution', fontsize=14, fontweight='bold')
        ax2.grid(True, alpha=0.3)
        
        # Add horizontal line for median correlation
        median_corr = np.median([f[1] for f in feature_importance])
        ax2.axhline(y=median_corr, color='#C00000', linestyle='--', alpha=0.7, 
                   label=f'Median: {median_corr:.3f}')
        ax2.legend()
        
        plt.tight_layout()
        plt.show()
        
        print("\nTop 10 Most Important Features (by absolute correlation):")
        print("-" * 70)
        print(f"{'Feature':<35} | {'Correlation':<12} | {'Direction':<10}")
        print("-" * 70)
        for feature, abs_corr, corr in feature_importance[:10]:
            direction = "Positive" if corr > 0 else "Negative"
            print(f"{feature:<35} | {corr:11.4f}  | {direction:<10}")
        
        # Save feature importance results
        importance_df = pd.DataFrame(feature_importance, columns=['Feature', 'Abs_Correlation', 'Correlation'])
        importance_df.to_csv('../results/feature_importance_ecommerce.csv', index=False)
        print(f"\nFeature importance results saved to: ../results/feature_importance_ecommerce.csv")
    else:
        print("No features available for importance analysis.")
else:
    print("Target variable 'class' not found in the dataset.")

# %% [markdown]
# ## 4. Credit Card Data Feature Engineering

# %%
print("\n" + "="*70)
print("CREDIT CARD DATA FEATURE ENGINEERING")
print("="*70)

# %% [markdown]
# ### 4.1 Original Data Overview

# %%
print("Original Features in Credit Card Data:")
print("-" * 50)
print(f"1. Time: Time elapsed since first transaction")
print(f"2. V1-V28: PCA-transformed features (anonymized)")
print(f"3. Amount: Transaction amount")
print(f"4. Class: Target variable (0=Non-Fraud, 1=Fraud)")

print(f"\nDataset shape: {cc_clean.shape}")
print(f"Fraud rate: {(cc_clean['Class'].mean() * 100):.4f}%")

# Display sample
print("\nSample of original data (first 3 rows):")
display(cc_clean.head(3))

# %% [markdown]
# ### 4.2 Create Credit Card Specific Features

# %%
print("\nCreating credit card specific features...")

# Create credit card features
cc_features = feature_engineer.prepare_creditcard_features(cc_clean)

# Display new features
new_cc_features = [col for col in cc_features.columns if col not in cc_clean.columns]
print(f"\nCreated {len(new_cc_features)} credit card specific features:")
for i, feat in enumerate(new_cc_features, 1):
    print(f"{i:2}. {feat}")

# Visualize new features
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# 1. Hour of day distribution
if 'hour_of_day' in cc_features.columns:
    axes[0, 0].hist(cc_features['hour_of_day'], bins=24, alpha=0.7, color='#2E75B6', edgecolor='black')
    axes[0, 0].set_title('Transaction Hour of Day', fontweight='bold')
    axes[0, 0].set_xlabel('Hour (0-23)')
    axes[0, 0].set_ylabel('Frequency')
    axes[0, 0].grid(True, alpha=0.3)
else:
    axes[0, 0].text(0.5, 0.5, 'Hour of day\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[0, 0].set_title('Transaction Hour', fontweight='bold')
    axes[0, 0].axis('off')

# 2. Day of week distribution
if 'day_of_week' in cc_features.columns:
    day_counts = cc_features['day_of_week'].value_counts().sort_index()
    day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    axes[0, 1].bar(day_names[:len(day_counts)], day_counts.values, color='#2E75B6', alpha=0.7)
    axes[0, 1].set_title('Transaction Day of Week', fontweight='bold')
    axes[0, 1].set_xlabel('Day')
    axes[0, 1].set_ylabel('Count')
    axes[0, 1].grid(True, alpha=0.3, axis='y')
else:
    axes[0, 1].text(0.5, 0.5, 'Day of week\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[0, 1].set_title('Transaction Day', fontweight='bold')
    axes[0, 1].axis('off')

# 3. Amount log distribution
if 'Amount_log' in cc_features.columns:
    axes[0, 2].hist(cc_features['Amount_log'], bins=50, alpha=0.7, color='#2E75B6', edgecolor='black')
    axes[0, 2].set_title('Log-Transformed Amount', fontweight='bold')
    axes[0, 2].set_xlabel('log(Amount + 1)')
    axes[0, 2].set_ylabel('Frequency')
    axes[0, 2].grid(True, alpha=0.3)
else:
    axes[0, 2].text(0.5, 0.5, 'Log amount\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[0, 2].set_title('Amount Transformation', fontweight='bold')
    axes[0, 2].axis('off')

# 4. V1 * Amount interaction
if all(col in cc_features.columns for col in ['V1', 'Amount_log']):
    axes[1, 0].scatter(cc_features['V1'], cc_features['Amount_log'], alpha=0.3, color='#2E75B6')
    axes[1, 0].set_title('V1 vs Log Amount', fontweight='bold')
    axes[1, 0].set_xlabel('V1 (PCA Feature)')
    axes[1, 0].set_ylabel('log(Amount + 1)')
    axes[1, 0].grid(True, alpha=0.3)
else:
    axes[1, 0].text(0.5, 0.5, 'V1 or Amount_log\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[1, 0].set_title('Feature Interaction', fontweight='bold')
    axes[1, 0].axis('off')

# 5. V statistics distribution
v_stats_cols = [col for col in ['V_mean', 'V_std', 'V_max', 'V_min'] if col in cc_features.columns]
if len(v_stats_cols) >= 2:
    v_stats_data = [cc_features[col].dropna() for col in v_stats_cols]
    axes[1, 1].boxplot(v_stats_data, labels=[col.replace('V_', '') for col in v_stats_cols], patch_artist=True)
    axes[1, 1].set_title('V Feature Statistics Distribution', fontweight='bold')
    axes[1, 1].set_ylabel('Value')
    axes[1, 1].grid(True, alpha=0.3, axis='y')
    
    # Color the boxes
    colors = ['#2E75B6', '#4ECDC4', '#FF6B6B', '#96CEB4']
    for i, patch in enumerate(axes[1, 1].patches):
        if i < len(colors):
            patch.set_facecolor(colors[i])
            patch.set_alpha(0.7)
else:
    axes[1, 1].text(0.5, 0.5, 'V statistics\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[1, 1].set_title('V Statistics', fontweight='bold')
    axes[1, 1].axis('off')

# 6. Correlation of new features with target
if 'Class' in cc_features.columns and len(new_cc_features) > 0:
    new_features_corr = []
    for col in new_cc_features:
        if col in cc_features.columns:
            corr = cc_features[col].corr(cc_features['Class'])
            new_features_corr.append((col, corr))
    
    if new_features_corr:
        new_features_corr.sort(key=lambda x: abs(x[1]), reverse=True)
        features = [f[0] for f in new_features_corr]
        correlations = [f[1] for f in new_features_corr]
        
        axes[1, 2].bar(range(len(features)), correlations, color='#2E75B6', alpha=0.7)
        axes[1, 2].set_title('New Features Correlation with Target', fontweight='bold')
        axes[1, 2].set_xlabel('Feature Index')
        axes[1, 2].set_ylabel('Correlation')
        axes[1, 2].set_xticks(range(len(features)))
        axes[1, 2].set_xticklabels(range(1, len(features) + 1))
        axes[1, 2].grid(True, alpha=0.3, axis='y')
    else:
        axes[1, 2].text(0.5, 0.5, 'No correlation data\navailable', 
                       ha='center', va='center', fontsize=12)
        axes[1, 2].set_title('Feature Correlation', fontweight='bold')
        axes[1, 2].axis('off')
else:
    axes[1, 2].text(0.5, 0.5, 'Target or new features\nnot available', 
                   ha='center', va='center', fontsize=12)
    axes[1, 2].set_title('Feature Correlation', fontweight='bold')
    axes[1, 2].axis('off')

plt.tight_layout()
plt.suptitle('Credit Card Feature Engineering Results', fontsize=14, fontweight='bold', y=1.02)
plt.show()

# %% [markdown]
# ### 4.3 Feature Transformation for Credit Card Data

# %%
print("\nPerforming feature transformation for credit card data...")

# Define numerical columns for scaling
cc_numerical_cols = []
for col in cc_features.columns:
    if col not in ['Time', 'Amount', 'Class'] and not col.startswith('V'):
        if cc_features[col].dtype in [np.float64, np.int64]:
            cc_numerical_cols.append(col)

if 'Amount_log' in cc_features.columns:
    cc_numerical_cols.append('Amount_log')

print(f"Numerical columns to scale: {cc_numerical_cols}")

# Scale numerical features
cc_scaled = feature_engineer.scale_numerical_features(cc_features, cc_numerical_cols)

# Handle missing values
cc_final = feature_engineer.handle_missing_values(cc_scaled)

print(f"\nFinal credit card dataset shape: {cc_final.shape}")
print(f"Total features: {cc_final.shape[1]}")

# %% [markdown]
# ### 4.4 Credit Card Feature Correlation Analysis

# %%
print("\nAnalyzing credit card feature correlations...")

# Select features for correlation analysis
analysis_features = []
for col in ['V14', 'V17', 'V12', 'V10', 'V16', 'V7', 'Amount_log', 'V_mean', 'V_std', 'Class']:
    if col in cc_final.columns:
        analysis_features.append(col)

if len(analysis_features) > 1:
    corr_matrix_cc = cc_final[analysis_features].corr()
    
    # Plot correlation heatmap
    plt.figure(figsize=(12, 10))
    mask = np.triu(np.ones_like(corr_matrix_cc, dtype=bool))
    sns.heatmap(corr_matrix_cc, mask=mask, cmap='coolwarm', center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .8},
                annot=True, fmt='.2f', annot_kws={'size': 10})
    plt.title('Feature Correlation Heatmap - Credit Card Data (Key Features)', 
              fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
else:
    print("Insufficient features for correlation analysis.")

# Analyze correlation with target
if 'Class' in cc_final.columns:
    cc_feature_importance = []
    
    for col in cc_final.select_dtypes(include=[np.number]).columns:
        if col != 'Class':
            corr = cc_final[col].corr(cc_final['Class'])
            cc_feature_importance.append((col, abs(corr), corr))
    
    # Sort by absolute correlation
    cc_feature_importance.sort(key=lambda x: x[1], reverse=True)
    
    # Plot top features
    top_cc_features = cc_feature_importance[:15]
    if top_cc_features:
        cc_features_list = [f[0] for f in top_cc_features]
        cc_correlations = [f[2] for f in top_cc_features]
        
        fig, ax = plt.subplots(figsize=(12, 8))
        
        colors = ['#2E75B6' if x > 0 else '#C00000' for x in cc_correlations]
        y_pos = np.arange(len(cc_features_list))
        ax.barh(y_pos, cc_correlations, color=colors, alpha=0.7)
        ax.set_yticks(y_pos)
        ax.set_yticklabels(cc_features_list)
        ax.set_xlabel('Correlation with Fraud Target', fontsize=12)
        ax.set_title('Top 15 Features by Correlation with Fraud - Credit Card Data', 
                     fontsize=14, fontweight='bold')
        ax.grid(True, alpha=0.3, axis='x')
        
        # Add correlation values
        for i, (corr, feat) in enumerate(zip(cc_correlations, cc_features_list)):
            ax.text(corr + (0.01 if corr >= 0 else -0.01), i, 
                    f'{corr:.3f}', va='center',
                    color='black', fontweight='bold')
        
        plt.tight_layout()
        plt.show()
        
        print("\nTop 10 Most Important Features for Credit Card Fraud Detection:")
        print("-" * 70)
        print(f"{'Feature':<15} | {'Correlation':<12} | {'Direction':<10}")
        print("-" * 70)
        for feature, abs_corr, corr in cc_feature_importance[:10]:
            direction = "Positive" if corr > 0 else "Negative"
            print(f"{feature:<15} | {corr:11.4f}  | {direction:<10}")
        
        # Save feature importance results
        cc_importance_df = pd.DataFrame(cc_feature_importance, 
                                       columns=['Feature', 'Abs_Correlation', 'Correlation'])
        cc_importance_df.to_csv('../results/feature_importance_creditcard.csv', index=False)
        print(f"\nFeature importance results saved to: ../results/feature_importance_creditcard.csv")
    else:
        print("No features available for importance analysis.")
else:
    print("Target variable 'Class' not found in the dataset.")

# %% [markdown]
# ## 5. Save Processed Data

# %%
print("\n" + "="*70)
print("SAVING PROCESSED DATASETS")
print("="*70)

# Create directories if they don't exist
import os
os.makedirs('../data/processed', exist_ok=True)
os.makedirs('../results', exist_ok=True)

# Save e-commerce processed data
fraud_final.to_csv('../data/processed/ecommerce_fraud_processed.csv', index=False)
print(f" E-commerce data saved to: ../data/processed/ecommerce_fraud_processed.csv")
print(f"   Original shape: {fraud_with_country.shape}")
print(f"   Processed shape: {fraud_final.shape}")
print(f"   Features added: {fraud_final.shape[1] - fraud_with_country.shape[1]}")

# Save credit card processed data
cc_final.to_csv('../data/processed/creditcard_processed.csv', index=False)
print(f"\n Credit card data saved to: ../data/processed/creditcard_processed.csv")
print(f"   Original shape: {cc_clean.shape}")
print(f"   Processed shape: {cc_final.shape}")
print(f"   Features added: {cc_final.shape[1] - cc_clean.shape[1]}")

# Save feature summary
feature_summary = pd.DataFrame({
    'Dataset': ['E-commerce', 'Credit Card'],
    'Original_Features': [fraud_with_country.shape[1], cc_clean.shape[1]],
    'Processed_Features': [fraud_final.shape[1], cc_final.shape[1]],
    'New_Features': [fraud_final.shape[1] - fraud_with_country.shape[1], 
                    cc_final.shape[1] - cc_clean.shape[1]]
})
feature_summary.to_csv('../results/feature_engineering_summary.csv', index=False)
print(f"\n Feature engineering summary saved to: ../results/feature_engineering_summary.csv")

print("\n" + "="*70)
print("FEATURE ENGINEERING COMPLETED SUCCESSFULLY!")
print("="*70)
```
